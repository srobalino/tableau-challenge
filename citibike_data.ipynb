```python
import pandas as pd
import numpy
```

```python
#import july data
july_file = "data/2013-07 - Citi Bike trip data.csv"
july_df = pd.read_csv(july_file)
july_df.head()
```

```python
#drop all null values
july_drop_df=july_df.dropna(how="all")
july_drop_df
```

```python
#change birth year column name 
july_drop= july_drop_df.rename(columns={'birth year': 'birthyear'})
july_drop.head()
```

```python
#drop \N values 
july_drop= july_drop[july_drop.birthyear != "\\N"]
july_drop
```

```python
july_drop.to_csv('data/CleanData/013-07_CitiBiketrip.csv')
```
